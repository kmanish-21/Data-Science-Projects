```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y").reset_index()

# Define the graph function
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Use the function to plot
make_graph(tesla_data, "Tesla Stock Price Over Time")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="5y").reset_index()

# Define the graph function
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data['Date'], data['Close'], label='Close Price', color='purple')
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Use the function to plot GameStop data
make_graph(gme_data, "GameStop Stock Price Over Time")
```
